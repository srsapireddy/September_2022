```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
```

```python
# Create Data
x=['base','2','3','5','7']
y = np.random.randint(low=0, high=100, size=10)
y = np.array([99, 95.75, 96.41, 98.40,  98.31])
```

```python
# Simple Bar Plot
plt.bar(x,y)
plt.xlabel('Number of Regions')
plt.ylabel("Accuracies")
plt.title('Accuracies Bar Plot')
plt.show()
```

```python
# Change color of each bar
plt.bar(x,y, color=['firebrick', 'green', 'blue', 'black', 'red',])
plt.xlabel('Number of Regions')
plt.ylabel("Accuracies")
plt.title('Barplot with colored bars')
plt.show()
```
